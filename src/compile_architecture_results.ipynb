```python
#to illustrate effect of neural architecture on performance...


import glob
import pandas as pd


tune_results=glob.glob('/media/krillman/240GB_DATA/codes2/SSL_GCM/data/*/saved_models/*.csv')


SSL_GAN_name='optuna_SSL_GAN_OPTUNA_VERIFY'
VAT_results_name='optuna_VIRTUAL_ADVER_VAT_OPTUNA_VERIFY_compiled_results'
#
# tune_results
```

```python
ssl_results=[s for s in tune_results if SSL_GAN_name in s]
ssl_results=[s for s in ssl_results if 'run1' not in s]

ssl_results




pdr_list=[pd.read_csv(s) for s in ssl_results]



pdr_list[-1]



for pdr in pdr_list:


    pdr['new_idx']=[f'{hs},{nh}' for hs,nh in zip(pdr['params_hidden_size'],pdr['params_nhidden'])]
    


#print(len(pdr_list))

for pdr,ssl_name in zip(pdr_list,ssl_results):

    dset_string=ssl_name.split('dataset_')[1].split('/saved_models/')[0]
    
    #print(dset_string)
    
    pdr['params_d_n']=dset_string
    
#print(len(pdr_list))



    
    
new_pdr=[]

for pdr in pdr_list:
    
    pdr=pdr[['value_optimal_acc','params_d_n','unlabelled_pc_acc','new_idx']]
    
    if pdr.shape[0]==5:
        print(pdr)
    
    new_pdr.append(pdr)
    
    
#print(len(new_pdr))




    
    #pdr=pdr[['value_optimal_acc','params_d_n','unlabelled_pc_acc','new_idx']]
    







all_pdr=pd.concat(new_pdr,ignore_index=True)

ssl_compiled=all_pdr



unique_data_names=ssl_compiled['params_d_n'].unique()
unique_arch=ssl_compiled['new_idx'].unique()


split_vat_compiled=[ssl_compiled[ssl_compiled.params_d_n==u] for u in unique_data_names]
split_vat_arch=[ssl_compiled[ssl_compiled.new_idx==u] for u in unique_arch]


orders=[]
for dn in ['d2','d3','d4','d5','d6','d7']:
    for n_u in ['2000','5000','10000']:
        no=f'n36_gaussian_mixture_{dn}_{n_u}'
        orders.append(no)

        
# dict_order={n:k for k,n in enumerate(orders)}



# s = pd.Series(dict_order)

# print (s)

pdd_list=[]

for k,architecture in enumerate(unique_arch):

    pdd=split_vat_arch[k][['params_d_n','unlabelled_pc_acc']]


    pdd=pdd.sort_values('params_d_n')

    pdd=pdd.set_index('params_d_n').transpose()

    pdd=pdd.rename(columns={'n36_gaussian_mixture_d2':'n36_gaussian_mixture_d2_2000',
                        'n36_gaussian_mixture_d3':'n36_gaussian_mixture_d3_2000',
                        'n36_gaussian_mixture_d4':'n36_gaussian_mixture_d4_2000',
                        'n36_gaussian_mixture_d5':'n36_gaussian_mixture_d5_2000',
                        'n36_gaussian_mixture_d6':'n36_gaussian_mixture_d6_2000',
                        'n36_gaussian_mixture_d7':'n36_gaussian_mixture_d7_2000',
                        })


    pdd=pdd[orders]

    pdd['arch']=unique_arch[k]
    
    pdd_list.append(pdd)
    
    
pdd_cat=pd.concat(pdd_list,axis=0)
ssl_formatted=pdd_cat.set_index('arch',drop=True)


ssl_formatted



new_row=[2000,5000,10000]*6

new_row=pd.DataFrame(new_row).transpose()

new_row.columns=ssl_formatted.columns

new_row.index=['n']

new_row['arch']=['n']

ssl_formatted_f=pd.concat((new_row,ssl_formatted.reset_index(drop=False)),ignore_index=True)

ssl_formatted_f=ssl_formatted_f.set_index('arch')


new_columns=['CG2']*3 + ['CG3']*3 + ['CG4']*3 + ['CG5']*3 + ['CG6']*3 + ['CG7']*3# + ['CG2']*3

print(len(new_columns))


ssl_formatted_f.columns=new_columns


ssl_fn='ssl_example_architectures.tex'

with pd.option_context("max_colwidth", 1000):
    ssl_formatted_f.to_latex(ssl_fn,encoding='utf-8', escape=False,column_format='rcccccccccccccccccc',float_format="%.3f")
```
